```python
Argentina=df[df["Country"]=="Argentina"]
print(Argentina)
```

```python
fig, axes = plt.subplots(figsize=(12, 8))
plt.title("Omicron in Argentina")
sns.lineplot(x='Date', y='Percentage of Omicron', data=Argentina, ax=axes)
```

```python
features=["India", "Germany", "Canada", "Japan", "Russia", "Argentina"]
six=df[(df["Country"]=="India") | (df["Country"]=="Germany") | (df["Country"]=="Canada") | (df["Country"]=="Japan") | (df["Country"]=="Russia") | (df["Country"]=="Argentina")]
print(six)
```

```python
fig, axes = plt.subplots(figsize=(14, 8))
plt.title("Presentase of Omicron in India, Germany, Canada, Japan, Russia, Argentina")
sns.lineplot(x='Date', y='Percentage of Omicron', hue='Country', data=six, ax=axes)
```

```python
fig, axes = plt.subplots(figsize=(14, 8))
plt.title("Percentage of Omicron in different countries")
sns.lineplot(x='Date', y='Percentage of Omicron', hue='Country', data=df, ax=axes)
```
